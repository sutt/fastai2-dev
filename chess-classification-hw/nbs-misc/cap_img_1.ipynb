```python
import imutils
import cv2
from matplotlib import pyplot as plt
from app.clienthelpers import draw
from PIL import Image
```

```python
def snap(cam_num=1, b_show=True):
    
    cam_num = 1
    camera = cv2.VideoCapture(cam_num)
    ret, frame = camera.read()
    
    if not(ret): 
        print('snap failed')
        return
    
    frame_resize = imutils.resize(frame, width=600)
    frame_final = draw(frame_resize)
    frame_final = cv2.cvtColor(frame_final, cv2.COLOR_BGR2RGB)
    
    if b_show:plt.imshow(frame_final)
    
    return frame_final
```

```python
img = snap()
img = img[50:400,200:400,:]
img = Image.fromarray(img)
```

```python
fn = 'back-2'
img.save('app/data/' + fn + '.jpg')
```
